```python
# Imoprting Libreries:
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

```python
# Send get request to the webpage server to get the source code of the page:

url1="https://www.cars24.com/buy-used-car?f=fuelType%3Ain%3Apetrol&sort=bestmatch&serveWarrantyCount=true&gaId=884271084.1675791575&storeCityId=2378&pinId=400001"
page1=requests.get(url1)

# See the page1 content:

soup1=BeautifulSoup(page1.content)
brand=[]
for i in soup1.find_all("h2",class_="_2lmIw"):
    brand.append(i.text.split(" ")[1])

manufacturing_year=[]
for i in soup1.find_all("h2",class_="_2lmIw"):
    manufacturing_year.append(i.text.split(" ")[0])
    
model=[]
for i in soup1.find_all("h2",class_="_2lmIw"):
    model.append(i.text.split(" ")[2])
    
driven_km=[]
for i in soup1.find_all("ul",class_="_13yb6"):
    driven_km.append(i.text.split(" ")[0])
    
owner=[]
for i in soup1.find_all("ul",class_="_13yb6"):
    owner.append((i.text.split("km")[1]).split(" ")[0])
    
fuel=[]
for i in soup1.find_all("ul",class_="_13yb6"):
    fuel.append(i.text.split("Owner")[1][:6])
    
rto=[]
for i in soup1.find_all("ul",class_="_13yb6"):
    rto.append(i.text.split("Petrol")[1])
    
loc=[]
for i in soup1.find_all("span",class_="_3DYbK"):
    loc.append(i.text)

price=[]
for i in soup1.find_all("div",class_="_18ToE"):
    price.append(i.text.split("₹")[1])
    
cars_df1=pd.DataFrame({"Brand Name":brand,"Manufacturing Year":manufacturing_year,"Model":model,"Driven KM":driven_km,"Owner":owner,"Fuel Type":fuel,"RTO":rto,"Locetion":loc,"Price":price})
cars_df1
```

```python
cars_df1.to_csv("cars_df1")
```
